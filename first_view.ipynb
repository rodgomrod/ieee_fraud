```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
import datetime

from utils.schemas import *
```

```python
data_folder = 'input'
train_transaction = pd.read_csv(data_folder+'/train_transaction.csv', dtype = schema_train_transaction)
train_identity = pd.read_csv(data_folder+'/train_identity.csv', dtype = schema_train_identity)
test_transaction = pd.read_csv(data_folder+'/test_transaction.csv', dtype = schema_test_transaction)
test_identity = pd.read_csv(data_folder+'/test_identity.csv', dtype = schema_test_identity)
```

```python
train_transaction.head()
```

```python
train_identity.head()
```

```python
print(train_transaction.shape, train_identity.shape)
print(test_transaction.shape, test_identity.shape)
```

```python
train_identity['has_identity'] = 1
test_identity['has_identity'] = 1
```

```python
train = train_transaction.set_index('TransactionID').merge(train_identity.set_index('TransactionID'), how='left', left_index=True, right_index=True)
```

```python
test = test_transaction.set_index('TransactionID').merge(test_identity.set_index('TransactionID'), how='left', left_index=True, right_index=True)
```

```python
train.shape, test.shape
```

```python
train['has_identity'] = train['has_identity'].fillna(0)
test['has_identity'] = test['has_identity'].fillna(0)
```

```python
train.head()
```

```python
START_DATE = datetime.datetime.strptime('2017-12-01', '%Y-%m-%d')
```

```python
train["date"] = train['TransactionDT'].apply(lambda x: (START_DATE + datetime.timedelta(seconds=x)))
train['dayofweek'] = train['date'].dt.dayofweek
train['hour'] = train['date'].dt.hour
train['day'] = train['date'].dt.day
train['month'] = train['date'].dt.month

test["date"] = test['TransactionDT'].apply(lambda x: (START_DATE + datetime.timedelta(seconds=x)))
test['dayofweek'] = test['date'].dt.dayofweek
test['hour'] = test['date'].dt.hour
test['day'] = test['date'].dt.day
test['month'] = test['date'].dt.month
```

```python
train.head()
```

```python
train.to_csv(data_folder+'/train_merged_3.csv', header=True, index=None)
print('train guardado')
test.to_csv(data_folder+'/test_merged_3.csv', header=True, index=None)
print('test guardado')
```
